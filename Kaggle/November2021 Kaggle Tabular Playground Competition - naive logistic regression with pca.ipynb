```python
pipe = Pipeline([('scaler', StandardScaler()), ('pca', PCA()), ('logit', LogisticRegression(solver = 'saga'))])
pipe.fit(df, y)
```

```python
del df
df = pd.read_csv("../input/tabular-playground-series-nov-2021/test.csv")
```

```python
id = df.pop('id')
# df = pd.DataFrame(pca.transform(df))
# df = pca.transform(df)
res = pd.DataFrame(pipe.predict_proba(df))
```
